```python
from rfpretriever.RFPRetriever import RFPRetriever
import logging

logging.basicConfig(level=logging.INFO)

rfp = RFPRetriever(
    pdf_path='./rfpparser/RFP 2032H323 Amendment.pdf',
    verbose=True
)
```

```python
rfp.annotations
```

```python
annotationstemp = []
for k, ann in rfp.annotations.items():
    tempann = {}
    for key in ann:
        if ann[key]:
            tempann[key] = True
        else:
            tempann[key] = False
    annotationstemp.append(tempann)
rfp.annotations = annotationstemp
rfp._save_annotations()
```

```python
import nest_asyncio

nest_asyncio.apply()

retriever = rfp.retriever

nodes_with_scores = retriever.retrieve(
    "What are the page limits?"
)
```

```python
from IPython.display import display, Markdown

display(Markdown(nodes_with_scores[0].node.text))
```

```python
from llama_index.llms.openai import OpenAI

llm = OpenAI(model="gpt-4")

prompt = (
    "Based on the the excerpt from the RFP, answer the question:"
    "\n"
    "---------------------------------------------\n"
    "Excerpt:\n"
    "{excerpt}"
    "\n"
    "---------------------------------------------\n"
    "{question}"
    "\n"
    "---------------------------------------------\n"
    "Answer:"
    )
```

```python
question = "What are the page limits?"
nodes_with_scores = retriever.retrieve(
    question
)
display(Markdown(llm.complete(prompt.format(excerpt=nodes_with_scores[0].node.text, question=question)).text))
```

```python
question = "What are the formating requirements?"
nodes_with_scores = retriever.retrieve(
    question
)
display(Markdown(llm.complete(prompt.format(excerpt=nodes_with_scores[0].node.text, question=question)).text))
```
