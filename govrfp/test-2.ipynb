```python
from rfpretriever.RFPRetriever import RFPRetriever
import logging

logging.basicConfig(level=logging.INFO)

rfp = RFPRetriever(
    pdf_path='./rfpparser/N6134024R0013 UC12 RFP.pdf',
    verbose=True
)
```

```python
from IPython.display import display, Markdown
from llama_index.llms.openai import OpenAI

llm = OpenAI(model="gpt-4")

prompt = (
    "Based on the the excerpt from the RFP, answer the question:"
    "\n"
    "---------------------------------------------\n"
    "Excerpt:\n"
    "{excerpt}"
    "\n"
    "---------------------------------------------\n"
    "{question}"
    "\n"
    "---------------------------------------------\n"
    "Answer:"
    )
```

```python
question = "What are the formating requirements? Also what are the page limits?"
nodes_with_scores = rfp.retrieve(
    question
)
print(len(nodes_with_scores))
_excerpt = ''
i = 0
for node in nodes_with_scores:
    _excerpt += f"Evidence {i+1}:\n"
    _excerpt += node.node.text
    _excerpt += "\n"
    i += 1
display(Markdown(llm.complete(prompt.format(excerpt=_excerpt, question=question)).text))
```
