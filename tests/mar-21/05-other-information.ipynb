```python
from govrfp.rfpretriever.RFPRetriever import RFPRetriever
import logging

logging.basicConfig(level=logging.INFO)

rfp = RFPRetriever(
    pdf_path='tests/mar-21/egsearch.pdf',
    verbose=True
)
```

```python
from llama_index.llms.openai import OpenAI
llm = OpenAI(model="gpt-3.5-turbo")
```

```python
# important dates
date_chunk = rfp.chunks()[0]
dates = llm.complete("List the important dates in the RFP" + "\n The following is an excerpt from the RFP" + date_chunk['text']).text
# save it into a file
with open("tests/mar-21/05-important-dates.txt", "w") as file:
    file.write(dates)
```

```python
# deliverables
deliverable_chunk = """Table 4: Quote Formatting & Page Limit Information Volume Title I Technical Capability/Risk (Non-Price) – 18 Pages* (Details in Section 3.2.) II Past Performance (Non-Price)- 1 Page Per Reference- Total 3-5 
pages  III Price Quote – No Page Limit IV Contract Documentation – No Page Limit """

deliverables = llm.complete("List all the deliverables in the RFP and the file names and format and page limit, if any" + "\n The following is an excerpt from the RFP:\n" + deliverable_chunk).text
# save it into a file
with open("tests/mar-21/05-deliverables.txt", "w") as file:
    file.write(deliverables)
print(deliverables)
```
